```python
def HEATWAVE_AREACOVER(path,timecat):
    """path: dir HW.p pickle file 
       timecat : 1,3,6,12  to match SPI(SPI1,SPI3,SPI6,SPI12)
    """

    import pickle
    
    hw1922 = pickle.load( open( path+"hw1921.p", "rb" ) )
    ahw1922 = pickle.load( open( path+"ahw1921.p", "rb" ) )
    HW1922 = pickle.load( open( path+"HW1921.p", "rb" ) )
    ahw1922[0].shape

    nlat,nlon = ahw1922[0].shape

    ty = 494-455 

    hw3,hw2,hw1,nohw=[],[],[],[]
    area_hmax,area_m ={},{}


    seqhwNO,seqhw3,seqhw2,seqhw1 = [],[],[],[]

    AHT,area = [],[]
    #HT3_color.append('White')
    #time= arange(bsize-1,ty,bsize)
    #time = np.insert(time, 0, 0, axis=0)
    AHT_color,HW =[],[]
    #AHT_color.append('White')
    for e in range(ty):
        Nnan = len(np.where(np.isnan(ahw1922[e])==True)[0])
        hw3.append(len(np.where(ahw1922[e]>=7)[0])*100/((nlat*nlon)-Nnan))
        hw2.append(len(np.where((ahw1922[e]>=5)&(ahw1922[e]<7))[0])*100/((nlat*nlon)-Nnan))
        hw1.append(len(np.where((ahw1922[e]>=3)&(ahw1922[e]<5))[0])*100/((nlat*nlon)-Nnan))
        nohw.append(len(np.where(ahw1922[e]<3)[0])*100/((nlat*nlon)-Nnan))
        HW.append(np.nanmax(HW1922[e]))
        
        if timecat ==1:
            area_hmax[e]=(np.max([hw3[e],hw2[e],hw1[e],nohw[e]]))
            #print(area_hmax)
            if hw3[e]==area_hmax[e]:
                AHT.append('black')
                #area.append(area_hmax[e])
                seqhw3.append(area_hmax[e])
                seqhw2.append(0)
                seqhw1.append(0)
                seqhwNO.append(0)
            elif hw2[e]==area_hmax[e]:
                AHT.append('gray')
                #area.append(area_hmax[e])
                seqhw2.append(area_hmax[e])
                seqhw3.append(0)
                seqhw1.append(0)
                seqhwNO.append(0)
            elif hw1[e]==area_hmax[e]:
                AHT.append('lightgray')
                #area.append(area_hmax[e])
                seqhw1.append(area_hmax[e])
                seqhw2.append(0)
                seqhw3.append(0)
                seqhwNO.append(0)
            elif nohw[e]==area_hmax[e]:
                AHT.append('white')
                #area.append(area_hmax[e])
                seqhwNO.append(area_hmax[e])
                seqhw2.append(0)
                seqhw1.append(0)
                seqhw3.append(0)

            AHT_color.append(AHT[e])
            
    return HW,AHT_color,hw1,hw2,hw3,nohw
```

```python
HW,AHT_color,seqhw1,seqhw2,seqhw3,seqhwNO = HEATWAVE_AREACOVER(path,1)
```
